```python
import pandas as pd
def_stat=pd.read_csv('RB_stat_def.csv').set_index('Rk')
#def_stat.set_index('Rk')
def_stat.head()
```

```python
def_stat[['First','Last']] = def_stat['Player'].str.split("\\",expand=True,)
def_stat.head()
#def_stat.to_csv('def_stat.csv',encoding = 'utf-8')
```

```python
ply_pos=pd.read_csv('Fifa20_data.csv')
ply_pos.head()
```

```python
ply_pos = ply_pos[['Name','Position']]
ply_pos.rename(columns={'Name':'First'},inplace=True)

```

```python
ply_pos.head()
```

```python
ply_pos=ply_pos[ply_pos['Position'].str.contains('RB',case=False)]
```

```python
ply_pos['Position']
```

```python
join_def_stat=pd.merge(def_stat,ply_pos,on='First')
```

```python
join_def_stat.head()
join_def_stat.sort_values(by=['90s'])
```

```python
join_def_stat=join_def_stat[join_def_stat['Nation'].str.contains('Eng',case=False)]
join_def_stat.sort_values(by=['90s'])
join_def_stat.head()
```

```python
total_rows=len(join_def_stat.axes[0])
total_rows
```

```python
join_def_stat = join_def_stat[['First','TklW','Succ','Blocks','Int']]
join_def_stat.set_index('First',inplace=True)
join_def_stat
```

```python
from soccerplots.radar_chart import Radar

params = list(join_def_stat.columns.values)

## range values
ranges = [(0.0, 5.0), (0.0,10.0), (0.06, 10.0), (0.0, 5.0)]

## parameter value
values = join_def_stat.loc['Trent Alexander-Arnold']
## instantiate object
radar = Radar()

## plot radar
fig, ax = radar.plot_radar(ranges=ranges, params=params, values=values, 
                                 radar_color=['#B6282F', '#FFFFFF'])
```
