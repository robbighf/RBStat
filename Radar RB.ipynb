### Upload Possesion Stat

```python
import pandas as pd
pos_stat=pd.read_csv('pos_stat.csv').set_index('Rk')
pos_stat
```

### Seperate Name Player

```python
pos_stat[['First','Last']] = pos_stat['Player'].str.split("\\",expand=True,)
pos_stat.head()
#def_stat.to_csv('def_stat.csv',encoding = 'utf-8')
```

```python
pos_stat=pos_stat[['First','Squad','Touches','Succ']]
pos_stat.rename(columns={'Succ':'SuccDribl'},inplace=True)
```

```python
pos_stat.head()
```

### Import Defensive Stat

```python
import pandas as pd
def_stat=pd.read_csv('RB_stat_def.csv').set_index('Rk')
#def_stat.set_index('Rk')
def_stat.head()
```

```python
def_stat[['First','Last']] = def_stat['Player'].str.split("\\",expand=True,)
def_stat.head()
#def_stat.to_csv('def_stat.csv',encoding = 'utf-8')
```

```python
def_stat = def_stat[['First','Squad','Nation','Pos','TklW','Succ','Blocks','Int']]
def_stat
```

### Import Passing Stat

```python
pass_stat=pd.read_csv('pass_stat.csv').set_index('Rk')
pass_stat
```

```python
pass_stat[['First','Last']] = pass_stat['Player'].str.split("\\",expand=True,)
pass_stat.head()
#def_stat.to_csv('def_stat.csv',encoding = 'utf-8')
```

```python
pass_stat = pass_stat[['First','Squad','KP','Cmp','PPA','CrsPA']]
pass_stat
```

### Join Table

```python
join_stat=pd.merge(def_stat,pos_stat,on=['First','Squad'])
join_stat
```

```python
join_stat=pd.merge(join_stat,pass_stat,on=['First','Squad'])
join_stat
```

### Read Data Player Position

```python
ply_pos=pd.read_csv('Fifa20_data.csv')
ply_pos.head()
```

### Take only Name and Position

```python
ply_pos = ply_pos[['Name','Position']]
ply_pos.rename(columns={'Name':'First'},inplace=True)
```

### Take Only Right Back Position

```python
ply_pos=ply_pos[ply_pos['Position'].str.contains('RB',case=False)]

ply_pos.to_csv('ply_pos.csv',encoding = 'utf-8')
```

### Merge Data Stat and Data Position

```python
join_pos_stat=pd.merge(join_stat,ply_pos,on='First')
join_pos_stat=join_pos_stat[join_pos_stat['Nation'].str.contains('ENG',case=False)]
join_pos_stat
```

```python
join_pos_stat=join_pos_stat[['First','TklW','Succ','Blocks','Int','Touches','SuccDribl','KP','Cmp','PPA','CrsPA']]
join_pos_stat.set_index('First',inplace=True)
join_pos_stat
```

### Take out data duplicate

```python
is_duplicate=join_pos_stat.index.duplicated(keep ='first') 
not_duplicate = ~is_duplicate
join_pos_stat = join_pos_stat[not_duplicate]
join_pos_stat
```
